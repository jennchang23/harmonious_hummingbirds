```python
import pandas as pd
import os
```

```python
#CSV file location
csv_path = "Arrest_Data_from_2010_to_Present.csv"
```

```python
#Reading the CSV file
arrest_data = pd.read_csv(csv_path, engine="python")
arrest_data
```

```python
#Determining number of of arrests for each charge type/description
arrest_data["Charge Description"].value_counts()
```

```python
#Creating a new CSV file that show us all of the unique values in "charge description" column
pd.DataFrame({"descriptions":arrest_data["Charge Description"].unique()}).to_csv("drunkdesc.csv")
```

```python
#Pulling data only with key words applicable to intoxication (e.g. drunk, drinking, alcohol, etc.)
drunk_descriptions = [desc for desc in arrest_data["Charge Description"].unique() if "DRUNK" in str(desc) or "ALCOHOL" in str(desc) or "DRINKING" in str(desc) or "DUI" in str(desc) or "INTOXICATION" in str(desc)]
drunk_descriptions
```

```python
#Just determining how many line items there are in the data set
arrest_data["Report ID"].count()
```

```python
#Creating a data frame with only drunk related information
drunk_data = arrest_data[arrest_data['Charge Description'].isin(drunk_descriptions)]
#drunk_data["Charge Description"].unique()
drunk_data.head()
```
