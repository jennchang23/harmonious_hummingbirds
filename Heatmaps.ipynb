```python
import gmaps
import pandas as pd

# Google developer API key
from config import api_key

# Configure gmaps

gmaps.configure(api_key)
print(api_key)
```

```python
# Create df
df_latlong = pd.read_csv('../harmonious_hummingbirds/output_data.csv')
df_latlong.dropna()
df_latlong.head()
```

```python
# Store latitude and longitude in locations
bar_locations = df_latlong[["lat", "long"]]
print(bar_locations)

bar_locations.dtypes

```

```python
# Plot Heatmap
fig = gmaps.figure()

# Create heat layer
heat_layer = gmaps.heatmap_layer(bar_locations, dissipating=False, max_intensity=20, point_radius=0.008)
 
# Add layer
fig.add_layer(heat_layer)

# Display figure
fig
```
