```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
all_drunk_data = "Intoxication Data_all.csv"
all_arrest_data = "Arrest_Data_from_2010_to_Present.csv"
```

```python
drunk_data = pd.read_csv(all_drunk_data)
all_data = pd.read_csv(all_arrest_data)
```

```python
display(drunk_data, all_data)
```

```python
all_count = all_data['Report ID'].count()
all_count
```

```python
drunk_count = drunk_data['Report ID'].count()
drunk_count
```

```python
ratio = drunk_count / all_count
ratio
```

```python
all_min_drunk = all_count - drunk_count
all_min_drunk
```

```python
pie_sizes = [drunk_count, all_min_drunk]
labels = ["Intox. Related Crimes", "All Crimes"]
colors = ["gray", "coral"]
```

```python
plt.pie(pie_sizes, labels=labels, colors=colors,
        autopct="%1.0f%%", shadow=True, startangle=140)
```
